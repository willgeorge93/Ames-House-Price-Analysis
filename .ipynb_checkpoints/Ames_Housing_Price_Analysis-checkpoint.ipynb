This project uses the [Ames housing data recently made available on kaggle](https://www.kaggle.com/c/house-prices-advanced-regression-techniques).

## Estimating the value of homes from fixed characteristics.

---

An outline of the strategy:
1. Develop an algorithm to reliably estimate the value of residential houses based on *fixed* characteristics.
2. Identify characteristics of houses that the company can cost-effectively change/renovate with their construction team.
3. Evaluate the mean dollar value of different renovations.

Then we can use that to buy houses that are likely to sell for more than the cost of the purchase plus renovations.

Note: *changeable* characteristics are generally anything that can be modified without having to undergo major construction on the house.

---
**Objective Details**:

- Train a model on pre-2010 data and evaluate its performance on the 2010 houses.
- Characterise the model.

```python
def correlation_heat_map(df):
    corrs = df.corr()

    # Set the default matplotlib figure size:
    fig, ax = plt.subplots(figsize=(11, 7))

    # Generate a mask for the upper triangle (taken from seaborn example gallery)
    mask = np.zeros_like(corrs, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Plot the heatmap with seaborn.
    # Assign the matplotlib axis the function returns. This will let us resize the labels.
    ax = sns.heatmap(corrs, mask=mask, annot=True)

    # Resize the labels.
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14, rotation=30)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14, rotation=0)
    ax.set_ylim([df.shape[1], 0])
    # If you put plt.show() at the bottom, it prevents those useless printouts from matplotlib.
    return fig, ax
```

```python
pre_2010 = house_dum[pos_pre]
house_2010 = house_dum[pos_2010]

y_train = pre_2010.pop('SalePrice')
X_train = pre_2010

y_test = house_2010.pop('SalePrice')
X_test = house_2010

scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=pre_2010.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=house_2010.columns)

linear = LinearRegression()
linear.fit(X_train, y_train)
predictions_2010 = linear.predict(X_test)


train_score_lin = linear.score(X_train, y_train)
test_score_lin = linear.score(X_test, y_test)

print('Training Score:', train_score_lin)
print('Testing Score:', test_score_lin)

scores = cross_val_score(linear, X_train, y_train, cv=10)

print('Cross Validation Scores:',scores)
print('Mean Cross Val Scores:', scores.mean())

# Characterize your model. How well does it perform? What are the best estimates of price?
```

## Determine any value of *changeable* property characteristics unexplained by the *fixed* ones.

---

Now that the model estimates the price of a house based on its static characteristics, what are the costs/benefits of quality, condition, and renovations? i.e. let's evaluate the effect in dollars of the renovatable features.

The estimates will fit the following two specifications:
1. The estimates of effects will be in terms of dollars added or subtracted from the house value. 
2. The effects will be on the variance in price remaining from the first model.

The residuals from the first model (training and testing) represent the variance in price unexplained by the fixed characteristics. Of that variance in price remaining, how much of it can be explained by the easy-to-change aspects of the property?

#### Largest impacts on the predictions

The largest effect on swinging the predicted prices away from the prices predicted by the model based on the property's fixed variables are:

(Absolute Values of Coefficients)

1. Overall Quality : 4929.65
2. Wooden Shingle roofing material : 4248.26
3. Good Pool Quality & Condition : 3647.99
4. Overall Condition : 3138.57
5. Good Kitchen Quality : 1946.28
6. Brick Face Exterior : 1465.30


This seems logical since of the variables included in the model you could assume that the largest deterrent or attractant would be a house's overall quality and condition, external features such as roofing and exterior masonry materials and quality of centrepiece rooms such as the kitchen which would cost a lot more than a standard room to redevelop (not just a simple redecoration). Also despite the presence of a pool at only 7 of the houses in the set, it appears that a Pool with good quality and condition has a large effect on price in those that have one.

Negative coefficients are those that arise from over prediction of the Sale Price for a particular property within the fixed predictor model. Therefore, the coefficients suggest that the presence of a quality Swimming Pool in good condition could signify overprediction of the initial model's estimated Sale Price of the property. Similarly for a good Kitchen Quality. 

Conversely, houses with Exterior Brick, Wooden Shingle roofing and good Overall Quality and Condition suggest that the house may have been underpredicted by the first model. Relationships between these variables in this sense could be that if the house has a high quality kitchen and pool area one would expect a high Sale Price, but if the overall quality and condition of the house are very poor and the roofing material is sub-optimal, this may suggest that the house requires a lot of redevelopment work and when the house has expensive fittings that need replacing, it could cost a lot of money on top of the Sale Price and therefore may be sold for an accordingly adjusted price.

## What property characteristics predict an "abnormal" sale?

---

The `SaleCondition` feature indicates the circumstances of the house sale. From the data file, we can see that the possibilities are:

       Normal	Normal Sale
       Abnorml	Abnormal Sale -  trade, foreclosure, short sale
       AdjLand	Adjoining Land Purchase
       Alloca	Allocation - two linked properties with separate deeds, typically condo with a garage unit	
       Family	Sale between family members
       Partial	Home was not completed when last assessed (associated with New Homes)
       
As can be seen above, Abnormal Sales are characterised by those houses with the conditions of 'trade, foreclosure, short sale'.

This suggests that these houses are likely to be sold for well under their market value potential.

If we can reliably indicate what features, if any, predict "abnormal" sales (foreclosures, short sales, etc.), then we may be able to indentify early the larger profit sales and target them for pre-auction purchase at a very cheap price.

The aim of the following is to determine and adequately validate which features of a property indicate this type of sale.

Given that of the different types of sales, Abnormal sales' inner conditions can be assumed to be somewhat ambiguous and varying by nature, I would suggest that small variations in the predictor data could push a sale into the Abnormal category. Given also that the criteria for Abnormal sales is that the house was a trade, foreclosure or short sale, this may be inherently difficult to capture via predictors whereby the factors influencing the Abnormal sale are factors that are external from the scope of the dataset.

By external factors, what is meant is that the house's features e.g. size, number of fireplaces, number of rooms, garage size, location etc. do not factor into the reasons for a trade, foreclosure or short sale.
Reasons for trade, foreclosure or short sale are likely circumstantial for the owners of the properties, perhaps financial hardship or short sale for the purpose of raising quick capital.

```python
# correlation heat map for the abovementioned variables most likely to indicate Abnormal Sale Conditions

def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
```

#### Correlation Matrix for all Sale Conditions

```python
ab_vars = ['MSSubClass', 'BsmtCond', 'BsmtFinType1', 'HouseStyle', 'BsmtFinType2', 'HeatingQC', 'KitchenQual', 'Functional', 'GarageFinish', 'SaleType', 'SaleCondition']

matrix = []
for i in range(len(ab_vars)):
    row = []
    for j in range(len(ab_vars)):
        confusion = pd.crosstab(house_orig[ab_vars[i]], house_orig[ab_vars[j]])
        val = cramers_v(confusion.values)
        row.append(val)
    matrix.append(row)

correlation_matrix = pd.DataFrame(matrix, columns=ab_vars, index=ab_vars)
```

```python
def plot_f1_lines(figsize=(8,6),fontsize=16):
    '''Create f1-score level lines to be added to the precison-recall plot'''

    fig, ax = plt.subplots(figsize=figsize)
    
    # add lines of constant F1 scores
    
    for const in np.linspace(0.2,0.9,8):
        x_vals = np.linspace(0.001, 0.999, 100)
        y_vals = 1./(2./const-1./x_vals)
        ax.plot(x_vals[y_vals > 0], y_vals[y_vals > 0],
                 color='lightblue', ls='--', alpha=0.9)
        ax.set_ylim([0, 1])
        ax.annotate('f1={0:0.1f}'.format(const),
                     xy=(x_vals[-10], y_vals[-2]+0.0), fontsize=fontsize)

    return fig, ax
```

```python
fig, ax = plot_f1_lines()
skplt.metrics.plot_precision_recall(y_test, model.predict_proba(X_test), 
                       plot_micro=True, 
                       title_fontsize=20, text_fontsize=16, cmap='Blues', ax=ax)
ax.legend(loc=[1.1,0])
plt.show()
```
